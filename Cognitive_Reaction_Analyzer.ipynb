```python
import random
import time
import pandas as pd
import matplotlib.pyplot as plt

class ReactionTimeTask:
    def __init__(self, trials=5):
        self.trials = trials
        self.reaction_times = []  # List to store reaction times

    def start_task(self):
        print("Welcome to the Reaction Time Task!")
        print(f"You will complete {self.trials} trials. Try to click as fast as you can.")
        print("Press 'Enter' as quickly as possible when you see 'Go!'")

        for trial in range(1, self.trials + 1):
            # Random delay between 2 and 5 seconds
            delay = random.randint(2, 5)
            print(f"\nTrial {trial} - Wait for 'Go!'")
            time.sleep(delay)  # Wait for random time

            # Show "Go!" to simulate the signal
            print("\nGO!")
            start_time = time.time()  # Record the time when "GO!" is shown

            # Wait for user input (simulated by pressing enter key)
            input("Press Enter NOW!")
            reaction_time = time.time() - start_time  # Calculate reaction time
            self.reaction_times.append(reaction_time)  # Store reaction time

            print(f"Your reaction time for this trial is {reaction_time:.2f} seconds.")

        self.show_statistics()

    def show_statistics(self):
        """ Calculate and display statistics based on the collected data """
        print("\nTask complete!")
        avg_reaction_time = sum(self.reaction_times) / len(self.reaction_times) if self.reaction_times else 0
        print(f"Average Reaction Time: {avg_reaction_time:.2f} seconds")

        # Create a DataFrame to display the reaction times for each trial
        df = pd.DataFrame({
            'Trial': range(1, self.trials + 1),
            'Reaction Time (seconds)': self.reaction_times
        })
        print("\nReaction Times for each trial:")
        print(df)

        # Plot the reaction times for each trial
        plt.figure(figsize=(10, 6))
        plt.plot(df['Trial'], df['Reaction Time (seconds)'], marker='o', color='b', linestyle='-', linewidth=2, markersize=8)
        plt.title("Reaction Time per Trial", fontsize=16)
        plt.xlabel("Trial", fontsize=12)
        plt.ylabel("Reaction Time (seconds)", fontsize=12)
        plt.grid(True)
        plt.show()

# Instantiate and start the task with 5 trials
task = ReactionTimeTask(trials=5)
task.start_task()
```
